```python
#a mask allows us to focus only on the portions of the image that interests us.
#Bitwise operations are used to extract specific regions of interest from images by using masks.
#Masks can be created by performing thresholding on images.
```

```python
import cv2
import numpy as np
 
hsvFrame = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
result = image.copy()
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)


green_lower = np.array([25, 52, 72], np.uint8)
green_upper = np.array([102, 255, 255], np.uint8)


green_mask = cv2.inRange(hsvFrame, green_lower, green_upper)
result = cv2.bitwise_and(result, result, mask=green_mask)
cv2_imshow(green_mask)
 
```
